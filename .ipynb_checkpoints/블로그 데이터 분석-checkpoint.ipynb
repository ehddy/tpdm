```python
# Ï†ïÍ∑ú ÌëúÌòÑÏãù
# Ï†ïÍ∑ú ÌëúÌòÑÏãù Ìï®Ïàò Ï†ïÏùò
import re

# ÌäπÏàòÎ¨∏Ïûê Ï†úÍ±∞ 
def apply_regular_expression(text):
    hangul = re.compile('[^ „Ñ±-„Ö£ Í∞Ä-Ìû£]')  # ÌïúÍ∏Ä Ï∂îÏ∂ú Í∑úÏπô: ÎùÑÏñ¥ Ïì∞Í∏∞(1 Í∞ú)Î•º Ìè¨Ìï®Ìïú ÌïúÍ∏Ä
    result = hangul.sub('', text)  # ÏúÑÏóê ÏÑ§Ï†ïÌïú "hangul"Í∑úÏπôÏùÑ "text"Ïóê Ï†ÅÏö©(.sub)ÏãúÌÇ¥
    return result
```
