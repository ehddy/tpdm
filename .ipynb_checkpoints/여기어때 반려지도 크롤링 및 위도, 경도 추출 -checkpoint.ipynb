```python
import pandas as pd
```

```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup

driver = webdriver.Chrome(executable_path="/Users/dongyokim/main/jupyter /chromedriver/chromedriver")
driver.set_window_size(1200, 800)
```

```python
driver.get("https://www.goodchoice.kr/product/recommend/153")
driver.implicitly_wait(3)    
```

```python
req = driver.page_source
soup= BeautifulSoup(req, 'html.parser')
sleep_lists = soup.select('#poduct_list_area > ul > li')
```

```python
sleep_name = []
address_name = []
for i in range(len(sleep_lists)):
    name = sleep_lists[i].select_one('a > div > div.name > strong').text.strip()
    address = sleep_lists[i].select_one('a > div > div.name > p:nth-child(3)').text.strip() 
    sleep_name.append(name)
    address_name.append(address)
    print("{0}. name: {1}, address: {2}".format(i, name, address))
```

```python
data = pd.DataFrame(columns=['업체명','주소', '업체구분'])
```

```python
data['업체명'] = sleep_name
data['주소'] = address_name
data['업체구분'] = '숙박지'
```

```python
import googlemaps

gmaps_key = 'AIzaSyA8s_dFBFkQUKvrRJq95PvKkkL8dYY3FIQ'  # API설정할 때 얻은 key
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
latitude = []
longitude = []
address = []
for i in range(data.shape[0]):
    name = data['업체명'][i]
    geocode_result = gmaps.geocode((name), language='ko') # 
    try:
        lat  = geocode_result[0]["geometry"]["location"]["lat"] # 리스트에서 위도 추출
        lon = geocode_result[0]["geometry"]["location"]["lng"] # 리스트에서 경도 추출
        re = geocode_result[0]['formatted_address']
        print("{0}. name: {1} lat: {2}, log: {3}, address: {4}".format(i, name, lat, lon, re))
        latitude.append(lat)
        longitude.append(lon)
        address.append(re)
    except:
        latitude.append(0)
        longitude.append(0)
        address.append(0)
        print("{0}: {1} 실패".format(i, name))

  
```

```python
data['위도'] = latitude
data['경도'] = longitude
data['주소'] = address 
```

```python
data = data.loc[data['위도'] != 0, :]
data.reset_index(drop=True, inplace=True)
```
