```python
import pandas as pd
from selenium import webdriver
import time
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome(executable_path="/Users/dongyokim/main/jupyter /chromedriver/chromedriver")
driver.set_window_size(1200, 800)
```

```python

loc_list = ['서울', '강원', '경기', '경남', '경북', '광주', '대구', '대전', '세종', '울산', '인천', '부산', '전남', 
            '전북', '제주', '충남', '충북']

name = []
for i in loc_list:
    driver.get("https://www.diningcode.com/list?addr={0}&query=%EC%95%A0%EA%B2%AC%EB%8F%99%EB%B0%98".format(i))
    driver.implicitly_wait(3)    

    time.sleep(5)
    req = driver.page_source
    soup= BeautifulSoup(req, 'html.parser')
    lists = soup.select('div.PoiList ol div div.RHeader > div > div.InfoHeader > h2')
    
    for i in range(len(lists)):
        name.append(lists[i].text.split('.')[1])
        
```

```python
data = pd.DataFrame(columns=['업체명', '업체구분'])
```

```python
data['업체명'] = name 
data['업체구분'] = '식당 및 카페'
```

```python
# 중복제거 
data = data.drop_duplicates()
data.reset_index(drop=True, inplace=True)
```

```python
import googlemaps

gmaps_key = 'AIzaSyA8s_dFBFkQUKvrRJq95PvKkkL8dYY3FIQ'  # API설정할 때 얻은 key
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
latitude = []
longitude = []
address = []
for i in range(data.shape[0]):
    name = data['업체명'][i]
    geocode_result = gmaps.geocode((name), language='ko') # 
    try:
        lat  = geocode_result[0]["geometry"]["location"]["lat"] # 리스트에서 위도 추출
        lon = geocode_result[0]["geometry"]["location"]["lng"] # 리스트에서 경도 추출
        re = geocode_result[0]['formatted_address']
        print("{0}. name: {1} lat: {2}, log: {3}, address: {4}".format(i, name, lat, lon, re))
        latitude.append(lat)
        longitude.append(lon)
        address.append(re)
    except:
        latitude.append(0)
        longitude.append(0)
        address.append(0)
        print("{0}: {1} 실패".format(i, name))
```

```python
data['위도'] = latitude
data['경도'] = longitude
data['주소'] = address 
```

```python
data = data.loc[data['위도'] != 0, :]
data.reset_index(drop=True, inplace=True)
```
