```python
latitude = []
longitude = []
address = []
for i in range(data.shape[0]):
    name = data['업체명'][i]
    geocode_result = gmaps.geocode((name), language='ko') # 
    try:
        lat  = geocode_result[0]["geometry"]["location"]["lat"] # 리스트에서 위도 추출
        lon = geocode_result[0]["geometry"]["location"]["lng"] # 리스트에서 경도 추출
        re = geocode_result[0]['formatted_address']
        print("{0}. name: {1} lat: {2}, log: {3}, address: {4}".format(i, name, lat, lon, re))
        latitude.append(lat)
        longitude.append(lon)
        address.append(re)
    except:
        latitude.append(0)
        longitude.append(0)
        address.append(0)
        print("{0}: {1} 실패".format(i, name))

  
```

```python
data['위도'] = latitude
data['경도'] = longitude
data['주소'] = address 
```

```python
data = data.loc[data['위도'] != 0, :]
data.reset_index(drop=True, inplace=True)
```
